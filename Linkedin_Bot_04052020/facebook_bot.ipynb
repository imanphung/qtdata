```python
import os, random, sys, time
# from urllib.parse import urlparse
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd
from getpass import getpass
from selenium.webdriver.chrome.options import Options
```

```python
option = Options()
option.add_argument("--disable-infobars")
# option.add_argument("start-maximized")
option.add_argument("--disable-extensions")

# Pass the argument 1 to allow and 2 to block
option.add_experimental_option("prefs", { 
    "profile.default_content_setting_values.notifications": 1 
})
```

```python
def getPersonalizedMessger(row, customMessage):
    return customMessage

def accessAccount(browser, username, password):
    elementID = browser.find_element_by_id('email')
    elementID.send_keys(username)
    elementID = browser.find_element_by_id('pass')
    elementID.send_keys(password)
    elementID.submit()

def getID(full_link):
    facebook_id = full_link.split('.com/',1)[1]
    if '?id=' in facebook_id:
        facebook_id = facebook_id.split('?id=',1)[1]
    return facebook_id
    
def sendMessegerFacebook(browser, full_link, facebook_id, customMessage):
    browser.get(full_link)
    time.sleep(random.uniform(3, 7))
    browser.find_element_by_xpath('//a[@href="/messages/t/'+ facebook_id +'/"]').click()
    time.sleep(random.uniform(3, 7))
    try:
        elementID = browser.find_element_by_xpath("//*[contains(text(), 'Type a message...')]")
    except:
        elementID = browser.find_element_by_xpath("//*[contains(text(), 'Nhập tin nhắn...')]")
    elementID = elementID.find_element_by_xpath("..")
    elementID = elementID.find_element_by_xpath("..")
    elementID = elementID.find_element_by_class_name("_5rpb")
    elementID = elementID.find_element_by_tag_name('span')
    elementID.send_keys(customMessage)
    time.sleep(random.uniform(3, 7))
    try:
        browser.find_element_by_xpath("/html/body/div[1]/div[8]/div[1]/div/div/div[4]/div/div[1]/div/div/div/div/div/div[1]/div/div[2]/div[4]/div/ul/li[8]/a").click()
    except:
        browser.find_element_by_xpath("/html/body/div[1]/div[7]/div[1]/div/div/div[4]/div/div[1]/div/div/div/div/div/div[1]/div/div[2]/div[4]/div/ul/li[8]/a").click()
    time.sleep(random.uniform(3, 7))
    #Close chat windows
    browser.find_element_by_xpath("//*[contains(@aria-label, 'Close tab') ]").click()
    
```

```python
link = 'https://www.facebook.com/login'
username = "nhociu_becung@yahoo.com"
password = getpass()

customMessage = "Hello. This is a test message."
```

```python
#Open link on chrome
browser = webdriver.Chrome(options=option,executable_path='driver/chromedriver.exe')
browser.get(link)
time.sleep(random.uniform(3, 7))

#Access account
accessAccount(browser, username, password)
time.sleep(random.uniform(3, 7))
print('Login to Facebook')

#Open file visitedUsers 
visitedUsers = pd.read_csv('facebook_visited_user.csv')

print('FacebookBot: START')
for _,row in data.iterrows():
    try:
        full_link = row['facebook_link']
        facebook_id = getID(full_link)
        if full_link in visitedUsers.values:
            print("Already sent to: ", full_link)
        else:
            sendMessegerFacebook(browser, full_link, facebook_id, customMessage)

            print(full_link, ': sent')
            visitedUsers = visitedUsers.append({'link':full_link, 'status':'done'}, ignore_index=True)

            # Otherwise, sleep to make sure everything loads
            time.sleep(random.uniform(3, 7))

    except:
        print(full_link, ': error')
        visitedUsers = visitedUsers.append({'link':full_link, 'status':'error'}, ignore_index=True)

# visitedUsers.to_csv('visited_user.csv', index=False)
print('FacebookBot: DONE!')
        
```
